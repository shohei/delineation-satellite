```python
import matplotlib.pyplot as plt
out = open('500epoch_log.txt').readlines()
```

```python
epoch = []
losses = []
for i,o in enumerate(out):
    try:
        loss = o.split('loss=')[1].split(',')[0]
        losses.append(float(loss))
        epoch.append(i)
    except Exception:
        pass
plt.plot(epoch,losses)
plt.savefig('loss.png')
```

```python
valid_per_image_ious = []
valid_dataset_ious = []
train_per_image_ious = []
train_dataset_ious = []

for i,o in enumerate(out):
    try:
        valid_per_image_iou = o.split('valid_per_image_iou=')[1].split(',')[0]
        valid_per_image_ious.append(float(valid_per_image_iou))        
        valid_dataset_iou = o.split('valid_dataset_iou=')[1].split(',')[0]
        valid_dataset_ious.append(float(valid_dataset_iou))        
        train_per_image_iou = o.split('train_per_image_iou=')[1].split(',')[0]
        train_per_image_ious.append(float(train_per_image_iou))        
        train_dataset_iou = o.split('train_dataset_iou=')[1].split(',')[0].split(']')[0]
        train_dataset_ious.append(float(train_dataset_iou)) 
    except Exception:
        pass
fig1 = plt.gcf()

plt.plot(valid_per_image_ious, label='valid_per_image_iou')
plt.plot(valid_dataset_ious, label='valid_dataset_iou')
plt.plot(train_per_image_ious, label='train_per_image_iou')
plt.plot(train_dataset_ious, label='train_dataset_iou')
plt.legend()

plt.show()
fig1.savefig('iou.png')
```
